```python
from ultralytics import YOLO
from matplotlib import pyplot as plt
from PIL import Image
```

```python
# Initialize your YOLO model (replace 'yolov8n.pt' with your model file)
my_new_model = YOLO("best.pt")

# Load your image
new_image = "11.jpg"

# Get predictions
new_results = my_new_model(new_image,show_labels=False,show_conf=False)

# Access bounding boxes (assuming new_results contains similar structure as 'results')
for r in new_results:
    im_array = r.plot()  # BGR numpy array of predictions
    im = Image.fromarray(im_array[..., ::-1])  # Convert to RGB PIL image
    im.show_labels=False
    im.save('results2.jpg')  # save image
    im.show()  # Show image
    im.save('results1.jpg')  # save image
```

```python
new_result_array = new_results[0].plot()
plt.figure(figsize=(12, 12))
plt.imshow(new_result_array)
```
